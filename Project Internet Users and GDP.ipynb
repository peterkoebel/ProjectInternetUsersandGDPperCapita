#Comparing the change of countries internet users by percent and GDP per capita

```python
import plotly as py
import plotly.graph_objs as go
import ipywidgets as widgets
import pandas as pd
```

```python
py.offline.init_notebook_mode(connected=True)
```

#Loading the Datasets

```python
data_iu = pd.read_csv('https://raw.githubusercontent.com/peterkoebel/ProjectInternetUsersandGDPperCapita/master/internet_users_pct.csv')
data_gdp = pd.read_csv('https://raw.githubusercontent.com/peterkoebel/ProjectInternetUsersandGDPperCapita/master/gdp_per_capita.csv')
data_con = pd.read_csv('https://raw.githubusercontent.com/peterkoebel/ProjectInternetUsersandGDPperCapita/master/continents.csv')
```

#Change the years column names to add iu_. This will distinguish them from the years in the GDP per capita dataset.

```python
iu_cols = list('iu_' + data_iu.columns[2:])
iu_cols.insert(0, 'Country Name')
iu_cols.insert(1, 'Country Code')
```

```python
data_iu.columns = iu_cols
```

```python
data_iu.head()
```

#Change the years column names to add gdp_. This will distinguish them from the years in the internet users by percent dataset.

```python
gdp_cols = list('gdp_' + data_gdp.columns[2:])
gdp_cols.insert(0, 'Country Name')
gdp_cols.insert(1, 'Country Code')
```

```python
data_gdp.columns = gdp_cols
```

```python
data_gdp.head()
```

#Merging the datasets.

```python
data = data_iu.merge(data_gdp)
```

```python
data = data.merge(data_con)
```

```python
data.head()
```

#Creating a yearly scatterplot with internet users by percent as the x-axis and GDP per capita as the y-axis.

```python
layout = go.Layout(title = 'Internet Users by % of population vs GDP per capita - Year 2002', yaxis = dict(title = 'GDP per Capita'), xaxis = dict(title = 'Internet Users by %'))
data_sp = go.Scatter(x = data['iu_2002'], y = data['gdp_2002'], mode='markers')
fig = go.Figure(data = [data_sp], layout = layout)
py.offline.iplot(fig)
```

#Improving the data visualization, with markers and colours.

```python
layout = go.Layout(title = 'Internet Users by % of population vs GDP per capita - Year 2002', yaxis = dict(title = 'GDP per Capita'), xaxis = dict(title = 'Internet Users by %'))
data_sp = go.Scatter(x = data['iu_2002'], y = data['gdp_2002'], mode='markers', text = data['Country Name'], marker=dict(size = 8, opacity = 0.5, color = data['Continent Color']))
fig = go.Figure(data = [data_sp], layout = layout)
py.offline.iplot(fig)
```

#Plotting for the year of 2016 for comparison.

```python
layout = go.Layout(title = 'Internet Users by % of population vs GDP per capita - Year 2016', yaxis = dict(title = 'GDP per Capita'), xaxis = dict(title = 'Internet Users by %'))
data_sp = go.Scatter(x = data['iu_2016'], y = data['gdp_2016'], mode='markers', text = data['Country Name'], marker=dict(size = 8, opacity = 0.5, color = data['Continent Color']))
fig = go.Figure(data = [data_sp], layout = layout)
py.offline.iplot(fig)
```

#In 2016 no Countries had less than 3.5% of their populaion using the internet, while GDP per capita didn't change much for those countries. Also many countries had over 80% of their population using the interet by 2016.

#Creating the ability to choose a year to plot.

```python
def choose_year(year):
    layout = go.Layout(title = 'Internet Users by % of population vs GDP per capita - Year {}'.format(year), yaxis = dict(title = 'GDP per Capita'), xaxis = dict(title = 'Internet Users by %'))
    data_sp = go.Scatter(x = data['iu_' + year], y = data['gdp_' + year], mode='markers', text = data['Country Name'], marker=dict(size = 8, opacity = 0.5, color = data['Continent Color']))
    fig = go.Figure(data = [data_sp], layout = layout)
    py.offline.iplot(fig)
```

```python
choose_year('2012')
```

#Creating a widget with a dropdown box with of all the available years to choose to plot. There isn't interesting scatterplots until 1990, when a higher than 0.1% of users started using the internet. In 1996 you start seeing multiple countries cross 10% of their population using the internet. In 2012, no country had less than 1% of their population using the internet. You click on a year, the use the up and down arrows on your keyboard to scroll through the years.

```python
year = widgets.Select(options=years_list, description ='Years')
widgets.interactive(choose_year, year = year)
```
